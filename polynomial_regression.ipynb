```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('/content/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```
print(X)
```

```
print(y)
```

#Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

#Splitting the dataset

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg2 = PolynomialFeatures(degree=2)
X_poly = poly_reg2.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

## Visualising the Linear Regression results

```
plt.scatter(X, y, color ='red')
plt.plot(X, lin_reg.predict(poly_reg2.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results of x to the power of 2

**This code snippet demonstrates the use of polynomial regression with the help of PolynomialFeatures from the sklearn.preprocessing module and LinearRegression from the sklearn.linear_model module.**

Let's break down the snippet step by step:

**from sklearn.preprocessing import PolynomialFeatures:**
This line imports the PolynomialFeatures class from the sklearn.preprocessing module.

PolynomialFeatures is a pre-processing step that transforms the original features into polynomial features, allowing us to perform polynomial regression.

**poly_reg = PolynomialFeatures(degree=2):**
This line creates an instance of the PolynomialFeatures class.
degree=2 specifies that we want to generate polynomial features up to the 2nd degree.
The degree parameter determines the maximum degree of the polynomial features to be generated.

**X_poly = poly_reg.fit_transform(X):**
This line applies the polynomial transformation to the input data X.
The fit_transform() method of PolynomialFeatures fits the transformer to the data and then transforms the data accordingly.
The result is stored in the X_poly variable, which contains the original features as well as the polynomial features.

**lin_reg_2 = LinearRegression():**
This line creates an instance of the LinearRegression class.
LinearRegression is a linear regression model used for fitting the polynomial regression line to the data.

**lin_reg_2.fit(X_poly, y):**
This line fits the polynomial regression model to the data.
The fit() method of LinearRegression takes the polynomial features X_poly and the target variable y as input.
It trains the model to learn the coefficients for the polynomial regression equation that best fits the data.
By using PolynomialFeatures to transform the original features into polynomial features and then fitting the LinearRegression model to the transformed data, the snippet performs polynomial regression. The lin_reg_2 model is then ready to make predictions based on the polynomial regression equation.


---
**Note**
In polynomial regression, we use X_poly as the input for lin_reg_2 because the polynomial features generated by PolynomialFeatures are used as the predictors (input variables) for the regression model.

After applying PolynomialFeatures to the original input data X, we obtain a new feature matrix X_poly that includes the original features as well as additional polynomial features. These polynomial features are created by taking different combinations and powers of the original features.

By fitting X_poly as the input and y as the target variable in the lin_reg_2.fit() method, we train the LinearRegression model to learn the coefficients for the polynomial regression equation that best fits the data. The model will consider the polynomial features when making predictions and estimating the relationship between the predictors and the target variable.

Using polynomial features allows us to capture non-linear relationships between the predictors and the target variable. The lin_reg_2 model, trained on the polynomial features, will be able to make predictions based on the learned polynomial regression equation.

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg2.predict(X_poly), color='blue')
plt.title('Truth or Bluf (Polynomial Regression to the power of 2)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results of X to the power of 4

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_2.predict(X_poly), color='blue')
plt.title('Truth or Bluf (Polynomial Regression to the power of 4)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

**This code snippet demonstrates the use of a polynomial regression model to fit a curve to data points and visualize the results using matplotlib**.

Let's break down the snippet step by step:

**X_grid = np.arange(min(X), max(X), 0.1):**
This line creates a new array X_grid using NumPy's arange() function.
The arange() function generates values from the minimum value of X to the maximum value of X, with a step size of 0.1.
This creates a fine-grained grid of values that will be used to plot the fitted curve later.

**X_grid = X_grid.reshape((len(X_grid), 1)):**
This line reshapes the X_grid array to have a single column.

It is necessary to reshape X_grid because the predict() method used later expects a 2D array as input.

**plt.scatter(X, y, color='red'):**
This line creates a scatter plot of the original data points using plt.scatter().
X represents the x-axis values, and y represents the corresponding y-axis values.
The data points are plotted as red dots.

**plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color='blue'):**
This line plots the fitted curve on the scatter plot.
X_grid is used as the x-axis values for the curve.
lin_reg_2.predict(poly_reg.fit_transform(X_grid)) calculates the predicted y-axis values for each value in X_grid.

The poly_reg.fit_transform() method transforms the X_grid values using polynomial features, allowing the curve to capture non-linear relationships.
The fitted curve is plotted as a blue line.

**plt.title('Truth or Bluff (Polynomial Regression)'):**
This line sets the title of the plot.

plt.xlabel('Position level') and plt.ylabel('Salary'): **bold text**
These lines set the labels for the x-axis and y-axis, respectively.

**plt.show():**
This line displays the plot on the screen.
By using the X_grid values to predict the corresponding y-axis values using the polynomial regression model, the snippet plots both the original data points and the fitted curve, allowing you to visualize the relationship between position level and salary.
